```python
# Source - https://stackoverflow.com/a
# Posted by xrisk
# Retrieved 2025-11-24, License - CC BY-SA 4.0

import pysrt
sub_dicts = []
for file_name in srt_files:
    sub_array = []
    file_path = 'srt files/' + file_name

    sub_raw = pysrt.open(file_path)
    for sub in sub_raw:
        sub_array.append([sub.start.to_time(), sub.text])
    sub_dicts.append({"title": file_name,
                     "subs": sub_array})

```

### Want to find the most common word in each film

```python
condensed_text = []
for movie in sub_dicts:

    all_film_text = ""
    for sub in movie["subs"]:
        all_film_text = all_film_text + " " + str.replace(str.replace(str.lower(sub[1]), '\n', ' '), "'", '')
    all_film_text = ' '.join([word for word in str.lower(all_film_text).split() if word not in (stopwords.words('english'))])
    
    condensed_text.append(all_film_text)
```

```python
for movie in condensed_text: 
    split_it = movie.split()
    Counters_found = Counter(split_it)
    most_occur = Counters_found.most_common(4)
print(most_occur)
```
